```python
import glob
import os
import numpy as np
%matplotlib notebook
from numpy import sqrt
import matplotlib.pyplot as plt
import scipy as sp
import math
import scipy

import scipy.io as io
import pdb
import scipy.signal as sig
import h5py
import signal
import csv

file_list = glob.glob("*.txt")

final_data={'rs1':[],'rs2':[],'rs3':[],'ls1':[],'ls2':[],'ls3':[],'us1':[],'us2':[],'us3':[],'ds1':[],'ds2':[],'ds3':[]}


d0arr=[]
d1arr=[]
d2arr=[]
r0arr=[]
r1arr=[]
r2arr=[]
l0arr=[]
l1arr=[]
l2arr=[]
u0arr=[]
u1arr=[]
u2arr=[]

d0arr=np.zeros((9,4,771))
d1arr=np.zeros((9,4,410))
d2arr=np.zeros((9,4,290))

u0arr=np.zeros((9,4,771))
u1arr=np.zeros((9,4,410))
u2arr=np.zeros((9,4,290))

l0arr=np.zeros((9,4,771))
l1arr=np.zeros((9,4,410))
l2arr=np.zeros((9,4,290))

r0arr=np.zeros((9,4,771))
r1arr=np.zeros((9,4,410))
r2arr=np.zeros((9,4,290))


arr=[]
for file_name in file_list:
    i=0
    i=0
    j=0
    array=[]
    c1=[]
    c2=[]
    c3=[]
    c4=[]
    timestamps=[]
    file=open(file_name,"r")
    for line in file:
        if i>=1280:
            array=line.split(",")
            timestamps.append(float(j)/256*1000)
            j=j+1
            c1.append(float(array[1]))
            c2.append(float(array[2]))
            c3.append(float(array[3]))
            c4.append(float(array[4]))
        i=i+1
    c1=scipy.signal.detrend(c1, axis=-1, type='linear', bp= [i for i in range(0,len(c1),128)])
    c2=scipy.signal.detrend(c2, axis=-1, type='linear', bp= [i for i in range(0,len(c1),128)])
    c3=scipy.signal.detrend(c3, axis=-1, type='linear', bp= [i for i in range(0,len(c1),128)])
    c4=scipy.signal.detrend(c4, axis=-1, type='linear', bp= [i for i in range(0,len(c1),128)])
    
    
    speed=int(file_name[9])-1
    
    trial=int(file_name[11])
    trial=trial+(3)*(int(file_name[7])-1)-1
        

    
    direction=file_name[6]
    if direction == 'r':
        dir=1
        if speed==0:
            c1=c1[0:771]
            c2=c2[0:771]
            c3=c3[0:771]
            c4=c4[0:771]
            r0arr[trial][0]=c1
            r0arr[trial][1]=c2
            r0arr[trial][2]=c3
            r0arr[trial][3]=c4
        if speed==1:
            c1=c1[0:410]
            c2=c2[0:410]
            c3=c3[0:410]
            c4=c4[0:410]
            r1arr[trial][0]=c1
            r1arr[trial][1]=c2
            r1arr[trial][2]=c3
            r1arr[trial][3]=c4
        if speed==2:
            c1=c1[0:290]
            c2=c2[0:290]
            c3=c3[0:290]
            c4=c4[0:290]
            r2arr[trial][0]=c1
            r2arr[trial][1]=c2
            r2arr[trial][2]=c3
            r2arr[trial][3]=c4
            
    elif direction =='d':
        dir=2
        if speed==0:
            c1=c1[0:771]
            c2=c2[0:771]
            c3=c3[0:771]
            c4=c4[0:771]
            d0arr[trial][0]=c1
            d0arr[trial][1]=c2
            d0arr[trial][2]=c3
            d0arr[trial][3]=c4

        if speed==1:
            c1=c1[0:410]
            c2=c2[0:410]
            c3=c3[0:410]
            c4=c4[0:410]
            d1arr[trial][0]=c1
            d1arr[trial][1]=c2
            d1arr[trial][2]=c3
            d1arr[trial][3]=c4
        if speed==2:
            c1=c1[0:290]
            c2=c2[0:290]
            c3=c3[0:290]
            c4=c4[0:290]
            d2arr[trial][0]=c1
            d2arr[trial][1]=c2
            d2arr[trial][2]=c3
            d2arr[trial][3]=c4
            
    if direction == 'l':
        dir=3
        if speed==0:
            c1=c1[0:771]
            c2=c2[0:771]
            c3=c3[0:771]
            c4=c4[0:771]
            l0arr[trial][0]=c1
            l0arr[trial][1]=c2
            l0arr[trial][2]=c3
            l0arr[trial][3]=c4
        if speed==1:
            c1=c1[0:410]
            c2=c2[0:410]
            c3=c3[0:410]
            c4=c4[0:410]
            l1arr[trial][0]=c1
            l1arr[trial][1]=c2
            l1arr[trial][2]=c3
            l1arr[trial][3]=c4
        if speed==2:
            c1=c1[0:290]
            c2=c2[0:290]
            c3=c3[0:290]
            c4=c4[0:290]
            l2arr[trial][0]=c1
            l2arr[trial][1]=c2
            l2arr[trial][2]=c3
            l2arr[trial][3]=c4
            
    elif direction =='u':
        dir=4
        if speed==0:
            c1=c1[0:771]
            c2=c2[0:771]
            c3=c3[0:771]
            c4=c4[0:771]
            u0arr[trial][0]=c1
            u0arr[trial][1]=c2
            u0arr[trial][2]=c3
            u0arr[trial][3]=c4
        if speed==1:
            c1=c1[0:410]
            c2=c2[0:410]
            c3=c3[0:410]
            c4=c4[0:410]
            u1arr[trial][0]=c1
            u1arr[trial][1]=c2
            u1arr[trial][2]=c3
            u1arr[trial][3]=c4
        if speed==2:
            c1=c1[0:290]
            c2=c2[0:290]
            c3=c3[0:290]
            c4=c4[0:290]
            u2arr[trial][0]=c1
            u2arr[trial][1]=c2
            u2arr[trial][2]=c3
            u2arr[trial][3]=c4
    writefile = open('csv/'+file_name+'.csv', "w")
    for index in range(len(c1)):
        writefile.write(str(c1[index]) + "," + str(c2[index]) +  ","+ str(c3[index]) + "," + str(c4[index]) + "\n")
    writefile.close()

            
final_data['rs1']=r0arr
final_data['rs2']=r1arr
final_data['rs3']=r2arr

final_data['ls1']=l0arr
final_data['ls2']=l1arr
final_data['ls3']=l2arr

final_data['us1']=u0arr
final_data['us2']=u1arr
final_data['us3']=u2arr

final_data['ds1']=d0arr
final_data['ds2']=d1arr
final_data['ds3']=d2arr

print(len(file_list))
    
```

```python
sp.io.savemat('final.mat',final_data)
from scipy.signal import butter, lfilter




def butter_bandpass_filter(data, lowcut, highcut, fs, order):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq

    b, a = butter(order, [low, high], btype='band')
    y = lfilter(b, a, data)
    return y

i=0

for key,value in final_data.items():
    for trial in range(0,9):
        for channel in range(0,4):
            value[trial][channel]=butter_bandpass_filter(value[trial][channel],0.5,30,256,5)
        
```
